```python
import requests

def scrape_jina_ai(url:str)->str:
    response = requests.get('https://r.jina.ai/'+url)
    return response.text
```

```python
def search(query):
    headers = {
        'Authorization': 'Bearer {}'.format(os.getenv('JINA_API_KEY'))
    }
    response = requests.get('https://s.jina.ai/'+query,headers=headers)
    return response
```

```python
respone = search('What happened to trump in the last few weeks')
```

```python
print(respone.text)
```

```python
with open('output.txt','w') as f:
    f.write(respone.text)
```

```python
from langchain_community.document_loaders import FireCrawlLoader
import os
loader = FireCrawlLoader(
    api_key=os.getenv("FIRECRAWL_API_KEY"), url="https://www.moneycontrol.com/india/stockpricequote/banks-private-sector/axisbank/AB16", mode="scrape"
)
```

```python
docs = loader.load()
```

```python
with open('output.txt','w') as f:
    f.write(docs[0].page_content)
```
